```python
# MBS cash flow table (Ex12-3)
import numpy_financial as npf
mrate=8.125/100
prate=7.5/100
pv=400000000
fv=0
for i in range(357):
    month=i+1
    mat=357-month+1
    if month <= 30:
        cpr=0.002*(month+3)
    else:
        cpr=0.06
    smm=1-pow((1-cpr),1/12)
    pmt=-npf.pmt(mrate/12, mat, pv, fv)
    mort_int=pv*mrate/12  #mortgage interest payment=smm*(pv-schedulepay)
    schedulepay=pmt-mort_int
    prepay=smm*(pv-schedulepay)
    pt_pay=schedulepay+prepay
    pt_int=prate*pv/12
    pt_cashflow=pt_pay+pt_int
    pv=pv-pt_pay
    print(month, round(smm,6), round(pmt,2), round(mort_int,2), round(pt_int,2), round(schedulepay,2), round(pt_pay,2), round(pt_cashflow,2), round(pv,2)) 
```

```python
# calculate average life (165PSA)
import numpy_financial as npf
mrate=8.125/100
prate=7.5/100
pv=400000000
fv=0
avglife=0
all1=0
for i in range(357):
    month=i+1
    mat=357-month+1
    if month <= 27:
        cpr=1.65*0.002*(month+3)
    else:
        cpr=0.06*1.65
    smm=1-pow((1-cpr),1/12)
    pmt=-npf.pmt(mrate/12, mat, pv, fv)
    mort_int=pv*mrate/12  #mortgage interest payment=smm*(pv-schedulepay)
    schedulepay=pmt-mort_int
    prepay=smm*(pv-schedulepay)
    pt_pay=schedulepay+prepay
    pt_int=prate*pv/12
    pt_cashflow=pt_pay+pt_int
    pv=pv-pt_pay
    avglife+=(pt_pay*month)/400000000/12
    all1+=pt_pay
print('Average life is',avglife)
all1
```
